```python
import pandas
import pickle
import json
from tqdm import tqdm_notebook
from pandas.io.json import json_normalize

data = []
for line in open('../Downloads/resumes.json', 'r'):
    data.append(json.loads(line))
```

```python
data = json_normalize(data)
```

```python
data.payment = pandas.to_numeric(data.payment, errors='coerce')
```

```python
tls = data[data.profession_tree_name  == "Разработчик"][data.payment > 150000]
```

```python
devs = data[data.profession_tree_name  == "Разработчик"][data.payment <= 150000][data.payment  >= 60000]
```

```python
newdevs = data[data.profession_tree_name  == "Разработчик"][data.payment  < 60000]
```

```python
institutes = ["" for i in range(len(newdevs))]
i = 0
for row in newdevs["base_education"]:
    e = json_normalize(row)
    for j in range(len(e)):
        institutes[i] += e.institute[j] + ', '
    if institutes[i] != "":
        institutes[i] = institutes[i][:-2]
    i += 1
```

```python
experiences = []
for row in newdevs["best"]:
    experiences.append(len(row))
experiences = [i//(max(experiences)//10) for i in experiences]
```

```python
languages = ["" for i in range(len(newdevs))]
i = 0
for row in newdevs["best"]:
    if ("java" in row.lower()):
        languages[i] += "java, "
    if ("c++" in row.lower()):
        languages[i] += "c/c++, "
    if ("python" in row.lower()):
        languages[i] += "python, "
    if ("1c" in row.lower()):
        languages[i] += "1c, "
    if ("c#" in row.lower()):
        languages[i] += "c#, "
    if ("haskell" in row.lower()):
        languages[i] += "haskell, "
    if (languages[i] != ""):
        languages[i] = languages[i][:-2]
    i += 1
```
